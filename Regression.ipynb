```python
x1=[3,4,5,6,2]
x2=[8,5,7,3,1]
y=[-3.7,3.5,2.5,11.5,5.7]
X1=sum(x1)/len(x1)
X2=sum(x2)/len(x2)
Y=sum(y)/len(y)
x12sum=0
x22sum=0
x1x2sum=0
x1ysum=0
x2ysum=0
for i in range(len(x1)):
    x12sum+=x1[i]**2
    x22sum+=x2[i]**2
    x1x2sum+=x1[i]*x2[i]
    x1ysum+=x1[i]*y[i]
    x2ysum+=x2[i]*y[i]

print(X1, X2, Y, x12sum, x22sum, x1x2sum, x1ysum, x2ysum)
```

```python
B1 = ((x22sum*x1ysum) - (x1x2sum*x2ysum))/((x12sum*x22sum)-(x1x2sum)**2)
```

```python
B2 = ((x12sum*x2ysum) - (x1x2sum*x1ysum))/((x12sum*x22sum)-(x1x2sum)**2)
```

```python
B0 = (Y- B1*X1-B2*X2)
```

```python
print(B1,B2,B0)
```

```python
x1=int(input("Enter x1: "))
x2=int(input("Enter x2: "))
y=B0+B1*x1+B2*x2
print(f"y predicted value for x1 = {x1} and x2 = {x2} = {y:.4f}")
```
